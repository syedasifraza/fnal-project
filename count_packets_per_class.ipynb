```python
import os
from scapy.all import rdpcap

def aggregate_packets_bytes(pcap_dir):
    """
    Aggregate the number of packets and total bytes across all PCAP files in a directory.
    """
    total_packets = 0
    total_bytes = 0
    pcap_files = [f for f in os.listdir(pcap_dir) if f.endswith('.pcapng')]

    for file in pcap_files:
        print(f"Processing {file}...")
        file_path = os.path.join(pcap_dir, file)
        try:
            packets = rdpcap(file_path)
            total_packets += len(packets)
            total_bytes += sum(len(pkt) for pkt in packets)
        except Exception as e:
            print(f"Error processing {file_path}: {e}")
    print(f"lala")
    return total_packets, total_bytes

# Define directories for each traffic type
streaming_dir = '/tank/swlarsen/2024_11_07/fcc-ht2'
transfer_dir = '/tank/swlarsen/2024_10_18/fcc-ht2/run1'
# Aggregate for each traffic type
streaming_packets, streaming_bytes = aggregate_packets_bytes(streaming_dir)
transfer_packets, transfer_bytes = aggregate_packets_bytes(transfer_dir)

# Print results
print("Streaming Traffic:")
print(f"Packets: {streaming_packets}, Bytes: {streaming_bytes}")
print("File Transfer Traffic:")
print(f"Packets: {transfer_packets}, Bytes: {transfer_bytes}")
import matplotlib.pyplot as plt

# Data for visualization
labels = ['Streaming', 'File Transfer']
packet_counts = [streaming_packets, transfer_packets]
byte_counts = [streaming_bytes, transfer_bytes]

# Bar Chart for Packets
plt.figure(figsize=(8, 5))
plt.bar(labels, packet_counts, color=['blue', 'orange'], alpha=0.7)
plt.title('Total Packets per Traffic Type')
plt.ylabel('Number of Packets')
plt.show()

# Bar Chart for Bytes
plt.figure(figsize=(8, 5))
plt.bar(labels, byte_counts, color=['blue', 'orange'], alpha=0.7)
plt.title('Total Bytes per Traffic Type')
plt.ylabel('Bytes')
plt.show()
```
